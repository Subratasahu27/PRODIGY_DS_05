```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
df=pd.read_csv("C:/Users/91740/OneDrive/Desktop/PRODIGY INTERNSHIP/TASK 5.csv")
df
```

```python
#Group the data by 'Sex of driver' and count the number of accident
accident_count=df['Sex_of_driver'].value_counts()

#Plot the bar chart 
accident_count.plot(kind='bar',color='blue')
#adding title and Labels 
plt.title('Accident Count by Sex of Driver')
plt.xlabel('Sex of Driver')
plt.ylabel('Number of Accidents')

#Show plot
plt.show()
```

```python
#Group The Data by 'Light condition','Road_Surface_type',and 'Weather condition'
grouped_light=df.groupby('Light_conditions').size()
grouped_road_surface=df.groupby('Road_surface_type').size()
grouped_weather=df.groupby('Weather_conditions').size()

#Create subplot
fix,axes=plt.subplots(nrows=1,ncols=3,figsize=(15,5))

#plotting the Light condition
grouped_light.plot(kind='bar',ax=axes[0],color='red')
axes[0].set_title('Road Accident by Light Condition')
axes[0].set_xlabel('Light Condition')
axes[0].set_ylabel('Number of Accidents')

#Plotting the Road surface
grouped_road_surface.plot(kind='bar',ax=axes[1],color='green')
axes[1].set_title('Road Accident by Road Surface')
axes[1].set_xlabel('Road Surface Type')
axes[1].set_ylabel('Number of Accidents')

#Plotting weather condition 
grouped_light.plot(kind='bar',ax=axes[2],color='pink')
axes[2].set_title('Road Accident by Weather Condition')
axes[2].set_xlabel('Weather Condition')
axes[2].set_ylabel('Number of Accidents')

#Adjust Layout
plt.tight_layout()

#Show plot
plt.show()



```

```python
#Group the data by 'type of condition' and count the number of accidents for each type of condition
collision_counts=df['Type_of_collision'].value_counts()

#Plotting the bar chart
plt.figure(figsize=(10,6))#Adjust the figure size if needed 
collision_counts.plot(kind='bar',color='skyblue')

#adding title and Labels 
plt.title('Distribution of Road Accident by Types of collision')
plt.xlabel('Type of Collision')
plt.ylabel('Number of Accidents')

#Rotate X-axis Labels for better readability(optional)
plt.xticks(rotation=45,ha='right')

#show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

pivot_table=df.pivot_table(index='Sex_of_driver',columns='Accident_severity',aggfunc='size',fill_value=0)
pivot_table
```

```python
#plotting the heat map with red and blue colors
plt.figure(figsize=(8,6))
sns.heatmap(pivot_table,annot=True,cmap='coolwarm',fmt='d')#'columns' cmap for red and blue colors

#Adding title and labels 
plt.title('Relationship between Sex of Driver and Accident severity')
plt.xlabel('Accident Severity')
plt.ylabel('Sex of Driver')

#show plot 
plt.tight_layout()
plt.show()
```
